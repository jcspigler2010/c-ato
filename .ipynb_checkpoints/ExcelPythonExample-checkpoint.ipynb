```python
# Control Vulnerability Description
# Security Control Number (NC/NA controls only)
# Office/Org
# Security Checks
# Resources Required
# Scheduled Completion Date
# Milestone with Completion Dates
# Milestone Changes
# Source Identifying Vulnerability 
# Status
# Comments
# Raw Severity
# Devices Affected
# Mitigations (in-house and in conjunction with the Navy CSSP)
# Predisposing Conditions
# Severity
# Relevance of Threat
# Threat Description
# Likelihood
# Impact
# Impact Description
# Residual Risk Level
# Recommendations
# Resulting Residual Risk after Proposed Mitigations

sheet = workbook.active

findings = {}

# Using the values_only because you want to return the cells' values
for col in sheet.iter_cols(min_row=8,
                           min_col=2,
                           max_col=25,
                           values_only=True):
    control = col[3]
    finding = {
        "Control Vulnerability Description": col[2],
        "Security Control Number (NC/NA controls only)": col[2],
        "Office/Org": col[3],
        "Security Checks": col[4],
        "Resources Required": col[5],
        "Scheduled Completion Date": col[6],
        "Milestone with Completion Dates": col[7],
        "Milestone Changes": col[8],
        "Source Identifying Vulnerability": col[9],
        "Status": col[10],
        "Comments": col[11],
        "Raw Severity": col[12],
        "Devices Affected": col[13],
        "Mitigations (in-house and in conjunction with the Navy CSSP)": col[14],
        "Predisposing Conditions": col[15],
        "Severity": col[16],
        "Relevance of Threat": col[17],
        "Threat Description": col[18],
        "Likelihood": col[19],
        "Impact": col[20],
        "Impact Description": col[21],
        "Residual Risk Level": col[22],
        "Recommendations": col[23],
        "Resulting Residual Risk after Proposed Mitigations": col[24]
    }
    findings[control] = finding

# Using json here to be able to format the output for displaying later
print(json.dumps(findings))
```
